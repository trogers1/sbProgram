```python
# finaldata = []
years = list(set(data.fy.values))
casc = list(set(data.casc.values))

finaldata = {"children":[{"children":[]}]}
for i in range(len(years)):
    yeardata = data[data.fy == years[i]]
    entry = {
        "year":years[i],
        "children":[]
    }
    proj_entry = {
        "type":"Projects",
        "children":[]
    }
    item_entry = {
        "type":"Items",
        "children":[]
    }

    for j in range(len(casc)):
        cascdf = yeardata[yeardata.casc == casc[j]]
        proj_entry2 = {
            "cascname":casc[j],
            "children":[]
        }
        
        item_entry2 = {
            "cascname":casc[j],
            "children":[]
        }
        
        project_title = list(set(cascdf.projectTitle.values))
        if len(project_title) > 0:
            for k in range(len(project_title)):
                d = cascdf[cascdf.projectTitle == project_title[k]]
                proj_entry3 = {
                    "project_name":project_title[k],
                    "size": list(set(d.projectSize))[0]
                }
                dd = d.item_size.values
                for l in range(len(dd)):
                    item_entry3 = {
                        "project_name":project_title[k],
                        "size": dd[l]
                    }

                proj_entry2["children"].append(proj_entry3)
                item_entry2["children"].append(item_entry3)
        else:
            proj_entry2["children"].append({"project_name": "", "size": 0})
            item_entry2["children"].append({"project_name": "", "size": 0})
            
        proj_entry["children"].append(proj_entry2)
        item_entry["children"].append(item_entry2)
    entry["children"].append(proj_entry)
    entry["children"].append(item_entry)
    finaldata["children"][0]["children"].append(entry)

# convert to pandas dataframe
finaldata_df = pd.DataFrame(finaldata)
finaldata_df.to_json('updated_data.json', orient = 'records')
finaldata

```
