```python
import sqlite3
import pandas as pd
import numpy as np
from pandas import DataFrame
```

```python
conn = sqlite3.connect('sbmacro.db')
```

```python
query1 = 'select name from sqlite_master where type = "table"'
q1_result = pd.read_sql_query(query1, conn)
q1_result
```

```python
items = pd.read_sql_query('select * from item', conn)
items.head(1)
```

items = pd.read_sql_query('select * from item', conn)
items.head(1)

```python
assoc = pd.read_sql_query('select * from assoc_item_sbfile', conn)
assoc.head(1)
```

```python
sbfile = pd.read_sql_query('select * from sb_file', conn)
sbfile.head(1)
```

```python
masterdetails = pd.read_sql_query('select * from master_details', conn)
```

```python
# STEP 1: connect masterdetails to items
item_ids = items.sb_id.values
masterdetails['item_id'] = masterdetails.sb_id.apply(lambda x: items[items.sb_id == x].id.values[0] if x in item_ids else '-')
masterdetails.head(1)

```

```python
# STEP 2: connect item id to sbfile id
assoc_ids = assoc.item_id.values
masterdetails['sbfile_id'] = masterdetails.item_id.apply(lambda x: assoc[assoc.item_id == x].sbfile_id.values[0] if x in assoc_ids else '-')
masterdetails.head(1)
```

```python
# STEP 3: retrieve file type from sb_file table
sb_file_id = sbfile.id.values
masterdetails['file_type'] = masterdetails.item_id.apply(lambda x: sbfile[sbfile.id == x].name.values[0].split('.')[-1] if x in sb_file_id else '-')
masterdetails.head(2)
```

```python
data = masterdetails[['fy', 'casc', 'file_type']]
data = data[data.file_type != '-']
```

```python
data.groupby(['casc', 'file_type']).count()
```

```python
finaldata = data.groupby(['casc', 'file_type']).count().reset_index()
```

```python
finaldata = finaldata.rename(columns = {'fy': 'total_count'})
```

```python
finaldata
```

```python
cascs = list(set(finaldata.casc))
data = []
for i in range(len(cascs)):
    data1 = {
        "casc":cascs[i],
        "children":[
            {
                "type": "",
                "children": [
                    {
                        "content": "",
                        "children": []
                    }
                ]
            }
        ]
    }
    matched_data = finaldata[finaldata.casc == cascs[i]]
    filetypes = matched_data.file_type.values
    filecount = matched_data.total_count.values
    for j in range(len(filetypes)):
        data2 = {
            "filetype":filetypes[j],
            "value":filecount[j]
        }
        data1["children"][0]["children"][0]["children"].append(data2)
    data.append(data1)
data
```

```python
# convert to pandas dataframe
finaldata_df = pd.DataFrame(data)
finaldata_df.to_json('treemap_data.json', orient = 'records')
```
